```python
offers = offer_retailer["brand_lookup"].astype(str).values
offers
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import numpy as np

def multiqa_mp_net_semantic_search(query, documents):
    # Initialize the model
    model = SentenceTransformer('multi-qa-mpnet-base-dot-v1')

    # Encode the documents and query
    doc_embeddings = model.encode(documents)
    query_embedding = model.encode([query])

    # Compute cosine similarity between the query and all documents
    similarities = cosine_similarity(query_embedding, doc_embeddings)

    # Create a dictionary to store each document with its similarity score
    similarity_dict = {doc: score for doc, score in zip(documents, similarities[0])}

    # Return the dictionary
    return similarity_dict
```

# BRAND TESTING

```python
brand_results = multiqa_mp_net_semantic_search("Barilla", offers)
brand_results
```

```python
# Sort the dictionary items by their values in descending order
sorted_results = sorted(brand_results.items(), key=lambda x: x[1], reverse=True)

# Get the top 5 results
top_5_results = sorted_results[:10]

# Print the top 5 results
for item, value in top_5_results:
    print(f"{item}: {value}")
```

```python
df = categories

# New dataframe to store the results
new_categories = pd.DataFrame(columns=["product_category", "sister_products"])

# Group by parent category and combine sister products
for parent_category, group in df.groupby("IS_CHILD_CATEGORY_TO"):
    sister_products = ", ".join(group["PRODUCT_CATEGORY"])
    new_categories = new_categories.append({"product_category": parent_category, "sister_products": sister_products}, ignore_index=True)

new_categories
```

#### when the input is from parent category

```python
# check if input_query is in a the column, then get all the child products
user_input = "Beverages"
search_items = []
if input_query in set(new_categories['product_category'].tolist()):
    
    # Find the response for the user input
    response_str = new_categories.loc[new_categories['product_category'].str.lower() == user_input.lower(), 'sister_products'].values[0]

    # Convert the response string to an array
    search_items = response_str.split(', ')

search_items   
```

```python
def top_5_results_for_item(item):
    search_category_offers = multiqa_mp_net_semantic_search(item, offers)
    sorted_results = sorted(search_category_offers.items(), key=lambda x: x[1], reverse=True)
    return sorted_results[:5]

return_suggestions = []
for item in search_items:
    item_dict = dict()
    item_dict["item"] = item
    item_dict["offers"] = top_5_results_for_item(item)
    print(item_dict)
```
